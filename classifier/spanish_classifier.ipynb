# Clasificador en español


Categorías del dataset de evaluación 1:
- 0: negativo
- 1: positivo
- 2: neutral
- 3: irrelevante

Se consideran tres clasificadores distintos. Uno para clasificar los tuits polarizados y los no polarizados, otro para los tuits negativos y positivos y finalmente uno que distinga los tres tuits directamente. En todos los casos se ha usado una SVM por ser el que mejores resultados arrojó de todos los que se consideraron pero también por poder manejar mejor el desbalanceo.



## Dos clases (polarizados y no polarizados)
Categorías:
- 0: Negativo y positivo
- 1: Neutral e irrelevante

Primero se juntan las dos bases de datos y se extraen todas las características:

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from scipy.sparse import hstack, csr_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn import neighbors, svm, tree, metrics, naive_bayes
```

```python
all_df = pd.read_csv('preprocessing/tweets_preprocessed.csv', index_col=0)
df = pd.read_csv('data/samples/evaluation1_labeled.csv', index_col=1)
df.labels = df.labels.map({0:0, 1:0, 2:1, 3:1})
df.rename(columns={'labels': 'label'}, inplace=True)
df.drop(columns=['Unnamed: 0'], inplace=True)

df2 = pd.read_csv('data/samples/vacunes_100rt_evaluation2_label.csv', index_col=0)
df2.label = df2.label.map({'neg': 0, 'neu': 1, 'irr': 1, 'pos': 0})

df = df.append(df2)

df.drop_duplicates(subset='content', keep='first', inplace=True)

all_df.loc[df.index, 'label'] = df.label

df_blob = pd.read_csv('data/EN/vacunes_100rt_en_textblob.csv', index_col=0)
all_df['polarity'] = df_blob.polarity
all_df['subjectivity'] = df_blob.subjectivity

df = all_df[all_df.label.notna()]
df.label = df.label.astype('int64')
#df.label = df.label.map({'neg': 0, 'neu': 1, 'irr': 1, 'pos':2})

more_features = pd.get_dummies(df.source)
for i, col in enumerate(more_features.columns):
    more_features.rename(columns={col: 'extra' + str(i)}, inplace=True)
df = pd.concat([df, more_features], axis=1)
df.outlinks = df.outlinks.apply(lambda x: 1 if len(x) > 2 else 0)

neg_pos_df = df[df.label == 0]
neu_df = df[df.label == 1]



df_train, df_test, y_train, y_test = train_test_split(df, df['label'], test_size=400, random_state=48451)
```

### Mejor clasificador

Se construye el clasificador con mejores parámetros del GridSearchCV.

```python
#Create a 10-fold cross validation set
acc = np.zeros((10,1))
f1 = np.zeros((10, 1))
precision = np.zeros((10, 1))
recall = np.zeros((10, 1))

conf_matrix = np.zeros((2,2))

kf=KFold(n_splits=10, shuffle=False) #random_state=0 has been removed
kf.get_n_splits()

i=0
for train_index, test_index in kf.split(df_train):
    #print(train_index)
    
    X_train, X_val = df_train.iloc[train_index].preprocessed_text, df_train.iloc[test_index].preprocessed_text
    bow_transformer = TfidfVectorizer(min_df=1, max_df=.4, ngram_range=(1,3)).fit(X_train)
    X_train = bow_transformer.transform(X_train)
    X_val = bow_transformer.transform(X_val)
    
    #extra_features_train = df_train.iloc[train_index][['polarity', 'subjectivity']].to_numpy() 
    #extra_features_val = df_train.iloc[test_index][['polarity', 'subjectivity']].to_numpy()

    #extra_features_train = df_train.iloc[train_index][['punctuation']].to_numpy() 
    #extra_features_val = df_train.iloc[test_index][['punctuation']].to_numpy()
    
    inherit = []
    inherit = [col for col in df_train.columns if col.startswith('extra')]
    inherit.extend( [col for col in df_train.columns if col.endswith('Count')] )
    
    inherit.append('outlinks')
    inherit.append('punctuation')
    inherit.append('polarity')
    inherit.append('subjectivity')
    

    extra_features_train = df_train.iloc[train_index][ inherit ].to_numpy()
    extra_features_val = df_train.iloc[test_index][ inherit ].to_numpy()
    
    scaler = MinMaxScaler().fit(extra_features_train)
    extra_features_train = csr_matrix ( scaler.transform(extra_features_train) )
    extra_features_val = csr_matrix ( scaler.transform(extra_features_val) )
    
    X_train = hstack((X_train, extra_features_train))
    X_val = hstack((X_val, extra_features_val))
    
    #y_train, y_val = y_train.iloc[train_index], y_train.iloc[test_index]
    y_train, y_val = df_train.iloc[train_index].label, df_train.iloc[test_index].label
    
    svc = svm.SVC(C=.1, kernel='rbf', class_weight="balanced", gamma='scale')
    #svc = naive_bayes.MultinomialNB()
    #svc = neighbors.KNeighborsClassifier(n_neighbors=1)
    #svc = tree.DecisionTreeClassifier()

    svc.fit(X_train,y_train)
    
    yhat_svc=svc.predict(X_val)
    
    acc[i][0] = metrics.accuracy_score(yhat_svc, y_val)
    f1[i][0] = metrics.f1_score(yhat_svc, y_val)
    precision[i][0] = metrics.precision_score(yhat_svc, y_val)
    recall[i][0] = metrics.recall_score(yhat_svc, y_val)
    
    conf_matrix += metrics.confusion_matrix(y_val, yhat_svc)
    
    i=i+1
    
plt.boxplot(acc);
for i in range(1):
    xderiv = (i+1)*np.ones(acc[:,i].shape)+(np.random.rand(10,)-0.5)*0.1
    plt.plot(xderiv,acc[:,i],'ro',alpha=0.3)
    
ax = plt.gca()
ax.set_xticklabels(['SVM'])   
ax.grid()

conf_matrix_normalized = np.apply_along_axis(lambda x: x/ x.sum(), 1, conf_matrix)
fig, ax = plt.subplots(1,2, figsize=(13,5))
sns.heatmap(conf_matrix, annot=True, ax=ax[0], cmap=sns.color_palette("Blues", as_cmap=True))
sns.heatmap(conf_matrix_normalized, vmin=0, vmax=1, center=.5, annot=True, ax=ax[1], cmap=sns.color_palette("Blues", as_cmap=True))

ax[0].set_title('Absolute values')
ax[1].set_title('Normalized values')
for ax_ in ax:
    ax_.set_xticklabels(['Negative and Positive', 'Neutral'])
    ax_.set_yticklabels(['Negative and Positive', 'Neutral'])
    ax_.set_xlabel('Predicted label')
    ax_.set_ylabel('True label')
```
